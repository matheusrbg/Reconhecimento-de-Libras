```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import os
```

```python
classes = {}
classes['train'] = {}
for folder in os.listdir("train"):
    classes['train'][folder] = len(os.listdir(f"train/{folder}"))
classes
```

```python
plt.bar(list(classes['train'].keys()), classes['train'].values())
plt.title("Distribuição de treinamento")
plt.savefig("train_dist.png")
```

```python
classes = {}
classes['val'] = {}
classes['test'] = {}
for folder in os.listdir("test"):
    classes['val'][folder] = len(os.listdir(f"test/{folder}"))//2
    classes['test'][folder] = len(os.listdir(f"test/{folder}")) - classes['val'][folder]
classes
```

```python
plt.bar(list(classes['test'].keys()), classes['test'].values())
plt.title("Distribuição de teste")
plt.savefig("test_dist.png")
```

```python
plt.bar(list(classes['val'].keys()), classes['val'].values())
plt.title("Distribuição de validação")
plt.savefig("val_dist.png")
```
